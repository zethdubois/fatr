$\text{NasaTLX} = \frac{\text{MD} + \text{TD} + (-\text{Performance}) + \text{Effort} + \text{Frustration Level}}{5}$

# Procedure Overview

1. **Consent and Signature Collection**

2. **Introduction and Familiarization (5-8 minutes)**
   - Basic nuclear power plants and Rancor simulator
   - Cyber threats, I&C malfunctions, and unnecessary shutdowns

3. **Practice Trial**
   - Guided simulator operation
   - Abnormal rapid plant shutdown procedure

4. **Experimental Trials (Block of Four)**
   - LOFW and SGTR scenarios
   - True Fault and Spoof conditions
   - Latin-square design

5. **Post-Trial Surveys**
   - NASA-TLX (after each trial)
   - Post-study survey (after final trial)
